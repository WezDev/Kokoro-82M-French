```python
# 2Ô∏è‚É£ Construire le mod√®le et charger le pack de voix par d√©faut
from models import build_model
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
MODEL = build_model('kokoro-v0_19.pth', device)
VOICE_NAME = [
    'af', # La voix par d√©faut est un m√©lange 50-50 de Bella et Sarah
    'af_bella', 'af_sarah', 'am_adam', 'am_michael',
    'bf_emma', 'bf_isabella', 'bm_george', 'bm_lewis',
    'af_nicole', 'af_sky',
][0]
VOICEPACK = torch.load(f'voices/{VOICE_NAME}.pt', weights_only=True).to(device)
print(f'Voix charg√©e: {VOICE_NAME}')
```

```python
# 3Ô∏è‚É£ Appeler la fonction generate, qui renvoie l'audio 24khz et les phon√®mes utilis√©s
from kokoro import generate
text = "How could I know? It's an unanswerable question. Like asking an unborn child if they'll lead a good life. They haven't even been born."
audio, out_ps = generate(MODEL, text, VOICEPACK, lang=VOICE_NAME[0])
# La langue est d√©termin√©e par la premi√®re lettre du VOICE_NAME:
# üá∫üá∏ 'a' => Anglais am√©ricain => en-us
# üá¨üáß 'b' => Anglais britannique => en-gb

# 4Ô∏è‚É£ Afficher l'audio 24khz et imprimer les phon√®mes de sortie
from IPython.display import display, Audio
display(Audio(data=audio, rate=24000, autoplay=True))
print(out_ps)
```
